# Codons & Protien Problem

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
def rnaToProtien(rnaLength): #Converts RNA to its respective Protien
  rna_rnd = np.random.choice(["A", "C", "U", "G"], size = rnaLength)
  rna = "".join(rna_rnd)
  #print(rna)
  pro_list =[]
  for place in np.arange(0,rnaLength, 3):
    codon = rna[place:place + 3]
    #print(codon)
    for row in range(len(df)):
      if codon == df.iloc[row,0]:
        a = str(df.iloc[row, 1])
        #print(a)
        pro_list.append(a)

  protien = "".join(pro_list)
  return protien
```

```python
rnaToProtien(6)
```

```python
def convert(mrna,place): #Converts mrna to trna and vice versa
  conv_mrna = []
  for pla in np.arange(place,len(mrna),3):
    if pla +3 > len(mrna):
      break
    codon = mrna[pla:pla + 3]
    codon_lis = list(codon)
    for letter in range(len(codon)):
      if codon[letter] == "U":
        codon_lis[letter] = "A"
      elif codon[letter] == "G":
        codon_lis[letter] = "C"
      elif codon[letter] == "A":
        codon_lis[letter] = "U"
      elif codon[letter] == "C":
        codon_lis[letter] = "G"

    codon = "".join(codon_lis)
    conv_mrna.append(codon)
  return "".join(conv_mrna)
```

```python
convert("UGACUG",0 )
```

```python
def rev(mrna, place): #Does the conversion backwards
  rev_mrna = mrna[::-1]
  return convert(rev_mrna, place)
```
